### Running a simulation

The function `run_simulation` below will setup a simulation of your robot on a given environment (`env`) and will use the policy `policy` you provide to control the robot, for a maximum of `max_steps` simulation steps. The previously created environments `env = Sidewalk(time_id, max_episode_time_step=1000, rendering=False)` determine the environment and whether the simulation is shown in a popup-window. The simulation function can also record all the (observation, actions) pairs (`record_data=True`).

The function signature of `run_simulation` is:
```python
rewards = run_simulation(policy, env, max_steps=500, verbose=1, record_data=False, delay=0.0, seed=-1, human_control=False)
    """ Run robot simulation
    Input arguments:
    - policy        # [function] the robot's policy function
    - env           # [instance of Sidewalk] the environment to simulate
    - record_data   # [True/False] if true, return all (observation, action) pairs from the simulation
    - max_steps     # [int] the maximum number of steps N to run the simulation
    - delay         # [float] a time delay that can be added to make the simulation run a bit slower
    - seed          # [int] location generation random seed - only set if >=0
    - verbose       # [int] how much text gets printed: 0 = none, 1 = final stats, 2 = all
    - human_control # [True/False] if true the key presses will be passed on the policy
    
    Returns:
    - rewards       # [numpy array of floats] all N rewards accumulated during the simulation
    - observations  # [numpy array N x H x W x 3] N observations, each observation being a WxH 3-channel image
    - actions       # [numpy array of ints] all N actions outputted by the given policy f
    
    Note: `observations` and `actions` are only returned if record_data=True
    """ 
```


````python
# user key input handler
from pyglet.window import key
import time

# keep track if which keys have been pressed in the popup window
# (will be used later for the human control)
KEY_PRESSED = {key.LEFT: False, key.RIGHT: False, key.UP: False}

# define the set of all actions    
ACTIONS = [0, 1, 2]
ACTION_NAMES = ['turn_left', 'turn_right', 'move_forward']
NUM_ACTIONS = 3 # number of distinct actions

def run_simulation(policy, env, max_steps=500, verbose=1, record_data=False, delay=0.0, seed=-1, human_control=False):
    """ Run robot simulation
    Input arguments:
    - policy        # [function] the robot's policy function
    - env           # [instance of Sidewalk] the environment to simulate
    - record_data   # [True/False] if true, return all (observation, action) pairs from the simulation
    - max_steps     # [int] the maximum number of steps N to run the simulation
    - delay         # [float] a time delay that can be added to make the simulation run a bit slower
    - seed          # [int] location generation random seed - only set if >=0
    - verbose       # [int] how much texts gets printed: 0 = none, 1 = final stats, 2 = all
    - human_control # [True/False] if true the key presses will be passed on the policy
    
    Returns:
    - rewards       # [numpy array of floats] all N rewards accumulated during the simulation
    - observations  # [numpy array N x H x W x 3] N observations, each observation being a WxH 3-channel image
    - actions       # [numpy array of ints] all N actions outputted by the given policy f
    
    Note: `observations` and `actions` are only returned if record_data=True
    """
    
    if verbose > 0:
        print(f'Starting simulation for {max_steps} steps.')
        if env.rendering:
            print('*** Press ESC key in popup window to stop the simulation! ***')
        print()

    if seed>=0:
        env.reset(seed=seed)
    else:
        env.reset(seed=int( time.time() * 1000.0 ))

    rewards = [] # will store the accumulated rewards
    observations = [] # will store the accumulated observations (only if record_data==True)
    actions = [] # will store the accumulated actions outputted by policy f (only if record_data==True)

    completed_steps = 0   
    
    
    try:
        # reset the simulation, and get the initial observation (robot "sensor measurement")
        obs = env.reinit(max_steps, wait_for_keypress=human_control)[0]
        
        # main simulation loop
        for step in range(max_steps):
            time.sleep(delay)
               
            if env.rendering:
                env.render()
            
            # get keyboard pressed button status from environment
            KEY_PRESSED.update(env.key_pressed)

            if env.stop_simulation: break
            
                
            # ** APPLYING YOUR POLICY **
            # execute the given policy on the observation to determine the robot's action
            action = policy(obs)
            
            # sanity check: is the policy implemented correctly?
            assert (isinstance(action, (int, np.integer))) # returned action should be a builtin or numpy integer
            assert (action in ACTIONS) # action should be an integer 0, 1, 2 or 3
            
            if verbose > 1:
                print(f'step {step}: action = {ACTION_NAMES[action]}')

            if record_data:
                # only store all the observation and action pairs during the simulation
                #   if the record_data argument is set to True
                observations.append(obs)
                actions.append(action)

            # execute simulation step with the given control input
            obs, reward, environment_done, _, info = env.step(action)
            completed_steps += 1
            
            if verbose > 1:
                print(f'step {step}: reward = {reward}')

            # collect all rewards in a list
            rewards.append(reward)

            # exit simulation when goal is reached
            if environment_done:
                break

    finally:
        # close the pop-up window,
        if env.rendering:
            env.close()
            env.window = None
        
    rewards = np.array(rewards)
    total_reward = np.sum(rewards)
    
    if verbose > 0:
        print(f'total reward after {completed_steps} steps: {total_reward}')
        print(f'average reward: {total_reward/completed_steps}')
    
    if record_data:
        return rewards, np.array(observations), np.array(actions, dtype=int)
    
    # by default, only return the rewards
    return rewards
````

The function `test_policy` below will run the `policy` in given environment (`env`) for a number `no_runs` different initial configurations. This function ensures that always the same set of configurations is used, allowing you to reproducibly compare the performance of different policies.

The function signature of `test_policy` is:
```python
no_completed, avg_reward = test_policy(policy, env, verbose=0, no_runs=10)
    """ Test a policy
    Input arguments:
    - policy        # [function] the robot's policy function
    - env           # [instance of Sidewalk] the environment to simulate
    - verbose       # [int] how much text gets printed: 0 = none, 1 = final stats, 2 = all
    - no_runs       # [int] how many configurations to evaluate
    
    Returns:
    - no_completed  # [int] number of runs in which the goal was reached
    - avg_reward    # [float] average reward over the runs
    """ 
```

````python
def test_policy(policy, env, verbose=0, no_runs=10):
    """ Test a policy
    Input arguments:
    - policy        # [function] the robot's policy function
    - env           # [instance of Sidewalk] the environment to simulate
    - verbose       # [int] how much text gets printed: 0 = none, 1 = final stats, 2 = all
    - no_runs       # [int] how many configurations to evaluate
    
    Returns:
    - no_completed  # [int] number of runs in which the goal was reached
    - avg_reward    # [float] average reward over the runs
    """ 
    
    no_completed = 0
    avg_reward = 0.0
  
    seeds = range(0,no_runs)
    run_count = 1
    for seed in seeds:
        print('Run %i / %i' % (run_count, no_runs))
        success = False
        run_count += 1
        rs = run_simulation(
            policy,
            env,
            record_data=False,
            verbose=verbose, 
            max_steps=300, 
            delay=0.0,
            seed=seed)

        avg_reward += np.mean(rs)
        if np.sum(rs) > 0:
            no_completed += 1
            success = True
            
        print('Success:', success)
    
    avg_reward /= float(no_runs)
    
    print()
    print(f'average reward: {avg_reward}')
    print(f'reached target: {no_completed} / {no_runs}')
    
    return no_completed, avg_reward
````

### Reward

To quantify how well a policy is working, the simulator will return the *rewards* that the robot collected at each simulation step. The rewards determine how well you are doing, and are based on whether or not you reach the target (red cube) or not and how quickly you do so:

* The robot receives a reward of 0 when not reaching the target within the given maximum number of steps.
* The robot receives a reward of 1 when reaching the target, at which time also the episode ends. The robot also gets a small malus for the number of steps it needs to reach the target, as a penalty for spending time. The equation for this is: 1.0 - 0.2 * (step_count / max_episode_steps)

Overall, the goal is to reach the red cube as quickly as possible.

 ### Clustering with K-Means
 
To reconstruct the clusters, we will begin by using k-means to try to differentiate the three class-labels. We choose k-means because if we have a lot of data k-means is computationally efficient (and therefore also scalable).

````python
from sklearn.cluster import KMeans

def feat_extract_clust_kmeans(observations,k):
    # Now, the shape of flattened_observation will be (522, 4800, 3) 
    flattened_observations = observations.reshape(observations.shape[0], -1 , observations.shape[3])
    # Calculate the mean along the second axis
    result_array = flattened_observations.mean(axis=1)
    # Now, the shape of flattened_observation will be (522, 3), perform k-means on dataset.
    kmeans = KMeans(n_clusters=k,random_state=1)
    labels = kmeans.fit_predict(result_array)
    return labels
````

### Clustering with Gaussian Mixture Model
In addition to k-means, we will introduce a different form of clustering: Gaussian Mixture. The reason for choosing the gaussian mixture is because unlike k-means, it provides a probability distribution over the clusters, which in turn allows for a more nuanced representation of uncertainty in data point clusters. The downside is that it is less computationally efficient.

Also, GMM clustering allows for non linear decision boundaries. This is especially important in this application as the clusters cannot be seperated by a linear classifier. 

````python
from sklearn.mixture import GaussianMixture

def feat_extract_clust(observations,k):
    flattened_observations = observations.reshape(observations.shape[0], -1 , observations.shape[3])
    result_array = flattened_observations.mean(axis=1)
    gmm = GaussianMixture(n_components=k, random_state=1)
    labels = gmm.fit_predict(result_array)
    return labels
````

## 2.2.2
### Propose a feature extraction method feat_extract which you will use in the subsequent sections to classify action, rather than time_ids.

For this section, we will propose two feature extraction methods. The first feature extraction is hand-crafted, and inspired by our feature extraction in step 1. The second one is a more common feature extraction, known as linear discriminant analysis (LDA). 



<u>Hand crafted feature extraction:</u>

In order to determine our move, we will try to locate the position of the package. (X,Y) where X is the shifted horizontal position to the package, and Y is our distance to the package. Based on the location of the package, it is possible for the model to determine whether we should move forward, left, or right. For example: If the package is in front of us the model should move forwards. If the package is to the left, it should rotate to the left. If the package cannot be found, the model should know that we can not move forwards, and should instead rotate left or right. 
Since the package has a very distinct red color, we can determine the package by looking at the intensity of red pixels in each column of the image. The extraction then continues by calculating the median red coordinate. This filters out the noise, and receives the center “X” coordinate of the package. Next we look at the “red intensity” at this coordinate. If it is near zero, we are most likely staring into the wall. If it is low, we are far away from the package, and if  it is high, we are close to the package. The intensity can be interpreted as the “Y” coordinate. We go from 60x80x3 (full image) features to only 2 (Position, Intensity)/(X,Y) 




<u>LDA Feature extraction:</u>

Linear discriminant analysis is a method that tries to find discriminant axes of the pixels that maximize separation between classes while at the same time minimizing the variance in each class. The number of discriminant axes is limited by n-1 where n=number of class labels. The (in our case) two axes create the largest class separation possible, and are therefore (by definition) also the two most informative features. 

One consideration to be made is that when we extract 14400 features down to two, we will obviously lose a lot of information along the way. In our hand-crafted feature extraction our model only looks at the package, and as a result will not have any information about the wall, or the road. This can lead to problems in the future, since these features are relevant when we test our model in the simulation. 


---

### Explain: Are there any important hyperparameters in your feature extractor?

<u>Hand-Crafted Feature-Extraction:</u>

The important hyperparameters for this feature extraction are our upper and lower bounds for how the red color is defined. The definition of “red intensity” is set by two hyperparameters in HSV color space that define the range we are looking for. 

<u>LDA:</u>

The most critical hyperparameter is n_components, which represents the dimension of the reduced feature space. In our case it can only be set to 1 or 2, since we have 3 class labels. We set this parameter to 2, because it will lead to a more informative feature space. 


---
## 2.2.4
### Explain: How will you decide on the values for these hyperparameters? What is the trade-off if this hyperparameter is either (too) low or (too) high?

<u>Hand-Crafted Feature-Extraction:</u>


By selecting an appropriate color range, the package-color can be included for all three time-ids, and it will not confuse other red objects (such as the wall) as a package. The reason we move from RGB -> HSV is because this makes the hyperparameters easier to tweak. If the space between the upper-bound and lower-bound is too big, our model will cause more noise, which can in turn interpret other red objects as the package, for example the wall during nighttime has a similar color to the package during daytime, and therefore we need to be careful setting these parameters.

<u>LDA:</u>

If the data is very high-dimensional, setting n_components to n_class_labels-1 might result in a dimensionality that results in a large computational complexity. Then it could be a good idea to lower n_components. If our n_components are too small however, we risk losing important information about the data. The choice of n_components is a trade off between information and computational complexity.

````python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

def feat_extract(observations, actions, n_components=2):
    lda = LinearDiscriminantAnalysis(n_components=n_components)
    N, H, W, C = observations.shape
    out = lda.fit_transform(observations.reshape(N, H * W * C), actions)
    return out, lda

def feat_extract_single(observation, lda):
    H, W, C = observation.shape
    observation = observation.reshape(1, H * W * C)
    return lda.transform(observation)

def plot_feat_extract(out, actions):
    plt.scatter(out[actions==0][0], out[actions==0][1])
    plt.scatter(out[actions==1][0], out[actions==1][1])
    plt.scatter(out[actions==2][0], out[actions==2][1])
    plt.show()
````

````python
import cv2

def percent_extract(image):
    # Convert the RGB image to HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

    # Define the hyperparameters for red color in HSV
    lower_red = np.array([0, 150, 150])
    upper_red = np.array([10, 255, 255])
 
    # Create a binary mask where pixels within the range are set to 1, and pixels outside the range are set to 0.
    mask = cv2.inRange(hsv_image, lower_red, upper_red)
    
    # Calculate the percentage of red pixels in each column
    red_pixel_percentage = np.mean(mask, axis=0)
    
    return red_pixel_percentage

def image_extract(image):
    # Calculate the percentage of red pixels in each column
    red_pixel_percentage = percent_extract(image)# np.mean(mask, axis=0)
    
    hyperparameter = 0
    binary_array = (red_pixel_percentage > hyperparameter).astype(int)
    
    indices = np.where(binary_array == 1)[0]
    
    if indices.size == 0:
        median_index = 0  # No occurrences of 1, so set median_index to 0
    else:
        median_index = int(np.median(indices))
    
    return np.array([median_index, red_pixel_percentage[median_index]])

def handcrafted_feat_extract(observations):
    # List to store red pixel percentages for each image
    out_array = np.zeros((len(observations), 2))

    for idx, image in enumerate(observations):
        # Perform image_extract for each image
        out = image_extract(image)
        
        # Append the red pixel percentages to the list
        out_array[idx] = out
        
    return out_array
````

<u>Why we choose to move forwards with LDA</u>

In order to choose what feature extraction to move on with, we should compare the advantages and limitations of our two choices.
Our handcrafted feature extraction is very tailored to the particular problem, and the features have a clear meaning where X represents our horizontal position, and Y represents our distance to the package. ### Convolutional Neural Network ( CNN )
For our first model, we decided to use a CNN. We built a wrapper class to handle initialisation, training and inference.

CNN models have a significant advantage over the other methods used in this network due to their ability to "understand meaning" in the input. As described in 3.1.1, CNNs can handle complex spatially varying data, such as images, much better than the other networks. We also believe that in reducing the feature space with LDA, some important application specific scene information will be lost, information we believe the CNN will be able to learn off of. 

Where the CNN underperforms however is in training, model size and inference time. CNNs and other neural networks take much more data to converge. This is problematic in that we only have about 500 samples in total, ~190 of which can be used for training out target day. This may be too little data for the model to converge or avoid overfitting. Training time is also an issue we are aware of. The not only does every forward and backward pass take considerably longer than more classical machine learning methods, but we will have to train over the data set a few times. This problem ties into the issue with inference time. A single pass through the CNN model takes significantly longer than the other methods, and this will be illustrated later in the notebook. 

````python
from keras.models import Sequential
from keras.layers import Conv2D, BatchNormalization, MaxPooling2D, Flatten, Dense, Softmax
from keras.utils import plot_model
from keras.optimizers import Adam
from keras.losses import CategoricalCrossentropy
from keras.metrics import CategoricalAccuracy

class CNN: 
    def __init__(self, input_shape=(None, 60, 80, 3), output_shape=3, num_conv_filters=[8, 16, 32], \
                        num_dense_neurons=[32, 64], optimizer=None, loss=None, lr=0.001):        
        self.input_shape = input_shape
        self.output_shape = output_shape
        self.optimizer = Adam(learning_rate=lr) if optimizer is None else optimizer(learning_rate=lr)
        self.loss = CategoricalCrossentropy() if loss is None else loss()
        
        self.model = Sequential([])
        for conv_filter in num_conv_filters:
            self.model.add(Conv2D(conv_filter, 3, strides=(1, 1), padding='same', activation='relu'))
            self.model.add(BatchNormalization())
            self.model.add(MaxPooling2D(2, 2))
        self.model.add(Flatten())
        for dense_neurons in num_dense_neurons:
            self.model.add(Dense(dense_neurons, activation='relu'))
        self.model.add(Dense(output_shape, activation=None))
        self.model.add(Softmax())
        
        self.build(input_shape)
    
    def build(self, input_shape):
        self.model.build(input_shape=self.input_shape)
        self.model.compile(optimizer=self.optimizer, loss=self.loss, \
              metrics=[CategoricalAccuracy()])
        #plot_model(self.model, to_file='cnn_model.png', show_shapes=True, show_layer_names=True)
        #print("CNN model built!")
    
    def predict(self, input_image, verbose=1):
        out = self.model.predict(input_image, verbose=verbose)
        return self.one_hot(out)
    
    def one_hot(self, model_out):
        arg_max = np.argmax(model_out, axis=1).reshape(-1)
        out = np.eye(self.output_shape)[arg_max]
        return out # output = np.float32

    def fit(self, X, y, epochs=1, verbose=1):
        y_1h = np.eye(self.output_shape)[y.reshape(-1)]
        return self.model.fit(X, y_1h, epochs=epochs, verbose=verbose)
    
    def evaluate(self, X, y, verbose=1):
        y_1h = np.eye(self.output_shape)[y.reshape(-1)]
        return self.model.evaluate(X, y_1h, verbose=verbose)
    
    def policy(self, observation):
        out = self.predict(observation.reshape(1, *observation.shape), verbose=0)
        return np.argmax(out)
````

## 3.1.2 (b)
### Dense Neural Network ( Dense NN )
The next model we decided to use was a dense neural network. We again built a wrapper class to handle initialisation, training and inference.

We believed that a dense neural network would  solve some of the issues seen with the CNN. Due to it's smaller architecture, we believed it would have a lower inference time to the convolutional network. Also due to the smaller feature space, we thought the network would take less time to train and reliably converge. Finally, we also think that in this setting, the dense NN's ability to understand non-linear and non-seperable data would be beneficial. We had a theory that in reducing the dimensionality of the feature space, there would be a significant amount of overlapping data points that would ultimately cause issue with the simpler machine learning methods. Our intuition was that the MLP would be able to better handle these situations when compared to the SVM and RF.

This model is still much larger than the classical machine learning methods however. The dense NN also does not have the ability to understand the scene as much as the convolutional neural network.

````python
class DenseNN: 
    def __init__(self, input_shape=(None, 2), output_shape=3, \
                        num_dense_neurons=[32, 64], optimizer=None, loss=None, lr=0.001):        
        self.input_shape = input_shape
        self.output_shape = output_shape
        self.optimizer = Adam(learning_rate=lr) if optimizer is None else optimizer(learning_rate=lr)
        self.loss = CategoricalCrossentropy() if loss is None else loss()
        
        self.model = Sequential([])
        for dense_neurons in num_dense_neurons:
            self.model.add(Dense(dense_neurons, activation='relu'))
        self.model.add(Dense(output_shape, activation=None))
        self.model.add(Softmax())
        
        self.build(input_shape)
    
    def build(self, input_shape):
        self.model.build(input_shape=self.input_shape)
        self.model.compile(optimizer=self.optimizer, loss=self.loss, \
              metrics=[CategoricalAccuracy()])
        #plot_model(self.model, to_file='dense_model.png', show_shapes=True, show_layer_names=True)
        #print("Dense NN built!")
    
    def predict(self, input_image, verbose=1):
        out = self.model.predict(input_image, verbose=verbose)
        return self.one_hot(out)
    
    def one_hot(self, model_out):
        arg_max = np.argmax(model_out, axis=1).reshape(-1)
        out = np.eye(self.output_shape)[arg_max]
        return out # output = np.float32

    def fit(self, X, y, epochs=1, verbose=1):
        y_1h = np.eye(self.output_shape)[y.reshape(-1)]
        return self.model.fit(X, y_1h, epochs=epochs, verbose=verbose)
    
    def evaluate(self, X, y, verbose=1):
        y_1h = np.eye(self.output_shape)[y.reshape(-1)]
        return self.model.evaluate(X, y_1h, verbose=verbose)
    
    def policy(self, observation):
        observation = image_extract(observation)
        out = self.predict(observation.reshape(1, *observation.shape), verbose=0)
        return np.argmax(out)
````

To help with fairly comparing the neural network output and the output of the classical machine learning methods, we will use the helper function below. 

````python
def one_hot_to_pred(pred):
    return np.argmax(pred, axis=1)
````

For the following section we create two functions per machine learning model. We provide an array of parameters we would like to explore to the ```XX_hp_exploration``` function. Inside the function, there's a for loop that iterated through every combination of parameters. For each combination, we perform a StratfiedKFold cross validation and save the mean and standard deviation of both the model accuracy and F1 score.

## 3.1.4 (a) 
### CNN Hyperparameter Tuning

````python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score

def train_and_evaluate_cnn_model(model, X_train, y_train, X_test, y_test, epochs=5):
    model.fit(X_train, y_train, epochs=epochs)
    y_pred = one_hot_to_pred(model.predict(X_test))
    print_accuracy_confusion(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='macro')
    print("f1 score : ", f1)
    print("-----------------------------")
    return accuracy_score(y_test, y_pred), f1

def cnn_hp_exploration(CNN, conv_layers, dense_layers, X, y, n_splits=5, epochs=5):
    metrics = []
    kf = StratifiedKFold(n_splits=n_splits)
    for conv_layer in conv_layers:
        for dense_layer in dense_layers:
            f1_scores = []
            accuracy = []
            for test, train in kf.split(X, y):
                cnn = CNN(num_conv_filters=conv_layer, num_dense_neurons=dense_layer)
                acc, f1 = train_and_evaluate_cnn_model(cnn, X[train], y[train], X[test], y[test], epochs=epochs)
                f1_scores.append(f1)
                accuracy.append(acc)
            metrics.append((np.mean(f1_scores), np.std(f1_scores), np.mean(accuracy), np.std(accuracy), conv_layer, dense_layer))
            
    return metrics
````

````python
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

def train_and_evaluate_svm_model(model, X_train, y_train, X_test, y_test):
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print_accuracy_confusion(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='macro')
    print("f1 score : ", f1)
    print("-----------------------------")
    return accuracy_score(y_test, y_pred), f1

def svm_hp_exploration(cs, X, y, n_splits=5):
    metrics = []
    kf = StratifiedKFold(n_splits=n_splits)
    for c in cs:
        print("Exploring c = ", c, "...")
        f1_scores = []
        accuracy = []
        for test, train in kf.split(X, y):
            svm = sklearn.svm.SVC(C=c)
            acc, f1 = train_and_evaluate_svm_model(svm, X[train], y[train], X[test], y[test])
            f1_scores.append(f1)
            accuracy.append(acc)
        metrics.append((np.mean(f1_scores), np.std(f1_scores), np.mean(accuracy), np.std(accuracy), c))
            
    return metrics
````

````python
def train_and_evaluate_rf_model(model, X_train, y_train, X_test, y_test):
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print_accuracy_confusion(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='macro')
    print("f1 score : ", f1)
    print("-----------------------------")
    return accuracy_score(y_test, y_pred), f1

def rf_hp_exploration(num_forests, max_depths, X, y, n_splits=5):
    metrics = []
    kf = StratifiedKFold(n_splits=n_splits)
    for num_forest in num_forests:
        for max_depth in max_depths:
            print("Exploring num_forest = ", num_forest, " & max_depth = ", max_depth)
            f1_scores = []
            accuracy = []
            for test, train in kf.split(X, y):
                rf = sklearn.ensemble.RandomForestClassifier(n_estimators=num_forest, max_depth=max_depth)
                acc, f1 = train_and_evaluate_rf_model(rf, X[train], y[train], X[test], y[test])
                f1_scores.append(f1)
                accuracy.append(acc)
            metrics.append((np.mean(f1_scores), np.std(f1_scores), np.mean(accuracy), np.std(accuracy), num_forest, max_depth))
            
    return metrics
````

As seen in the simulations above, a good accuracy score does not represent a good performance in the simulation. This is because the accuracy score is defined by how well our model fits the path of the training data, and not on how well our model is performing in the actual simulation. As previously mentioned, we introduced the f1 score to get a good balance between precision and recall. The f1 score essentially punishes the model if it has a bad performance in any of the minority classes, and this is why it is a better indicator than the accuracy score. 

The problem is however, that neither of these metrics represent how well the model is actually performing in a real life setting. Many of our models with both high accuracy score and f1 score are overfitting, and perform poorly in the simulation. In order to get a better representation of how well our model is actually performing in a real life-setting, we should use the data from the simulation to measure our performance.

## 3.2.1
### What are the most important hyperparameters of your chosen algorithm?

The hyperparameters to tune in our CNN are the following, ranked from most to least important:

     Dense Layers and number of neurons
The number of dense layers changes the architecture of the neural network, and as a result it directly affects the model's complexity and capacity. Adding more layers and neurons increases the complexity, but risks overfitting. Reducing them too much however, can make the model underfit. It's important to get the right tradeoff, and therefore this is one of the most important parameters to tune. 

     ConvLayers and number of filters
Tuning the number of convolutional layers and the number of filters, will affect how our model extracts features, like edges, the road, or the sky from our input image. Just like with the dense layers, it's a trade-off between complexity- and risk of overfitting.

     Loss function
The loss function measures how well our model performs, and it essentially defines what our optimization goal is. Changing the loss function impacts the network's ability to learn. 

     Learning Rate
The learning rate controls the step-size during the model-training. We want a high learning rate in order to converge fast to the minima, but it can't be too big because then we will overshoot the optimal solution. The learning rate affects the ability to learn effectively

     Optimizer
The optimizer influences how the model updates its weights during training, which in turn affects the training speed. This is not our most important hyperparameter, since all of our training speeds are already quite low. 

     Number of Epochs
Tuning the number of epochs is a trade-off between how long the model gets to learn and potentially overfitting. We want the model to learn as much as possible, but since we already have a lot of data, we don't need to set this parameter too high. We have already seen our other models having trouble with overfitting and don't want to create the same for our CNN. Although not the most important, it's still a hyperparameter that needs to be tweaked.

````python
from sklearn.model_selection import StratifiedKFold

EPOCHS = 20
NUM_RUNS = 15
NUM_SEARCHES = 15
NUM_SPLITS = 5

def train_and_evaluate_final_cnn_model(model, policy, X, y, epochs=5, num_runs=5):
    model.fit(X, y, epochs=epochs, verbose=0)
    cnn_num_compeleted, cnn_mean_reward = test_policy(policy, envs_norender[time_id], no_runs=num_runs, verbose=0)
    print("CNN Performance")
    print("    Number of runs completed : ", cnn_num_compeleted)
    print("    Mean Reward              : ", cnn_mean_reward)
    return cnn_num_compeleted, cnn_mean_reward

def final_cnn_hp_exploration(conv_layers, dense_layers, loss_funcs, optimizers, lrs, X, y, \
                                n_splits=NUM_SPLITS, n_searches=NUM_SEARCHES, epochs=EPOCHS, no_runs=NUM_RUNS):
    metrics = []
    for _ in range(n_searches):
        conv_layer = conv_layers[np.random.randint(len(conv_layers))]
        dense_layer = dense_layers[np.random.randint(len(dense_layers))]
        loss_func = np.random.choice(loss_funcs)
        optimizer = np.random.choice(optimizers)
        lr = np.random.choice(lrs)
        runs = []
        rewards = []
        print("------------------------------------------------")
        print("Conv Layers      : ", conv_layer)
        print("Dense Layers     : ", dense_layer)
        print("Loss Functon     : ", loss_func)
        print("Optimizer        : ", optimizer)
        print("Learning Rate    : ", lr)
        for _ in range(n_splits):
            cnn = CNN(num_conv_filters=conv_layer, num_dense_neurons=dense_layer, \
                      optimizer=optimizer, loss=loss_func, lr=lr)
            def test_cnn_policy(observation):
                out = cnn.predict(observation.reshape(1, *observation.shape), verbose=0)
                return np.argmax(out)
            run, reward = train_and_evaluate_final_cnn_model(cnn, test_cnn_policy, X, y, epochs=EPOCHS)
            runs.append(run)
            rewards.append(reward)
        metrics.append((np.mean(runs), np.std(runs), np.mean(rewards), np.std(rewards), \
                        loss_func, optimizer, conv_layer, dense_layer, lr))
            
    return metrics
````

## 4.1
### How can the 3 provided datasets be used to train a model that can generalize, and even more importantly how can they be used to evaluate whether a model can generalize? (Hint: Lecture 2)

In order to generalize the model we want it to be trained on a varied dataset representing all times of day. We used all of the data bar the final time as training data. We also shuffled the data before using it in our training loop.

To evaluate if the model can generalise we can perform n-fold cross validation. The n-fold cross validation splits our dataset into N-folds. We then perform N experiments, each time using a different fold as a test set, and the remaining folds as training sets. We can use this strategy to measure the performance where we look at the mean reward, mean accuracy, standard deviation of the reward, and standard deviation of the accuracy

````python
from sklearn.model_selection import KFold

def predict_classifier(cnn,observations,actions,n_folds,n_epochs,n_runs):
    #Shuffle the data
    permutation = np.random.permutation(len(observations))
    observations_shuffle = observations[permutation]
    actions_shuffle = actions[permutation]
    
    accuracy = np.array([])
    reward = np.array([])
    kf = KFold(n_folds, shuffle=False)
    for train_index, test_index in kf.split(observations_shuffle):
        X_train, X_test = observations_shuffle[train_index], observations_shuffle[test_index]
        y_train, y_test = actions_shuffle[train_index], actions_shuffle[test_index]
        
        cnn.fit(X_train, y_train, epochs=n_epochs)
        y_pred = one_hot_to_pred(cnn_model.predict(X_test))
        
        accuracy_value = accuracy_score(y_test, y_pred)
        accuracy = np.append(accuracy,accuracy_value)
        
        def cnn_kfold_policy(observations_shuffle):
            out = cnn.predict(observations_shuffle.reshape(1, *observations_shuffle.shape), verbose=0)
            return np.argmax(out)
        
        cnn_num_compeleted, cnn_mean_reward = test_policy(cnn_kfold_policy, envs_norender[1], no_runs=n_runs)
        reward= np.append(reward, cnn_mean_reward)
        
    
    mean_accuracy = np.mean(accuracy)
    std_accuracy = np.std(accuracy)
    mean_reward = np.mean(reward)
    std_reward = np.std(reward)
    
    return mean_reward, std_reward, mean_accuracy, std_accuracy
````

````python
n_folds = 3
n_epochs = 1
n_runs = 5

cnn_model = CNN(num_conv_filters=[6, 8, 12], num_dense_neurons=[32, 24])
mean_reward, std_reward, mean_accuracy, std_accuracy = predict_classifier(cnn_model,observations,actions,n_folds,n_epochs,n_runs)
print("The mean reward is: ", mean_reward)
print("The std deviation reward is: ", std_reward)
print("The mean accuracy is: ", mean_accuracy)
print("The std deviation accuracy  is: ", std_accuracy)
````

There are two methods we would like to perform in order to increase model robustness. The first one is to perform data augmentation, and the second one is to include dropout layers in our model architecture.

Data augmentation is the practice of applying transformations ( such as rotation, scaling, saturation changes, brightness changes, etc. ) to the training dataset in order for the model to have a more diverse and larger set of data to learn off of. This practice has two benefit for this application. Firstly, the augmentations will make our model more generalisable as it will train off of a more diverse dataset and increase the likelihood that the model will learn more meaningful features. The other benefit is that the model will have more training data to learn off of. This is especially important because we since we are using neural network. More data will likely have a correlation with model performance in simulation, given that overfitting does not happen.

The other change we are making to our model is the inclusion of dropout layers in the convolutional blocks. Dropout layers will randomly zero connections between layers. As described in section 4.2, dropout layers reduce layer interdependance and increases feature independance, which is especially important in understanding spatial data such as images.

## 4.5
### Implement at least one of those methods.

Below we will implement both generalisation methods described in section 4.4. We will create a new CNN wrapper class with dropout included in the convolution block. We will add the dropout layer before the convolutional layer, as described in this paper. This wrapper can be seen below.

We also added an data augmentor into our network. The augmentations include randomly changing the brightness and contrast in the images. We dont think it will be necessary to add any rotation, cropping or scaling augmentations to the data.

````python
from keras.layers import Dropout
from keras.layers import RandomContrast, RandomBrightness

class generalCNN: 
    def __init__(self, input_shape=(None, 60, 80, 3), output_shape=3, num_conv_filters=[8, 16, 32], \
                        num_dense_neurons=[32, 64], dropout_p=0.15, optimizer=None, loss=None, \
                         lr=0.001, augmentor=None, contrast_factor=0.2, brightness_factor=0.4):        
        self.input_shape = input_shape
        self.output_shape = output_shape
        self.optimizer = Adam(learning_rate=lr) if optimizer is None else optimizer(learning_rate=lr)
        self.loss = CategoricalCrossentropy() if loss is None else loss()
        if augmentor is None:
            augmentor = Sequential([
                RandomContrast(contrast_factor),
                RandomBrightness(brightness_factor)
            ])
        
        self.model = Sequential([])
        self.model.add(augmentor)
        for conv_filter in num_conv_filters:
            self.model.add(Dropout(dropout_p))
            self.model.add(Conv2D(conv_filter, 3, strides=(1, 1), padding='same', activation='relu'))
            self.model.add(BatchNormalization())
            self.model.add(MaxPooling2D(2, 2))
        self.model.add(Flatten())
        for dense_neurons in num_dense_neurons:
            self.model.add(Dense(dense_neurons, activation='relu'))
        self.model.add(Dense(output_shape, activation=None))
        self.model.add(Softmax())
        
        self.build(input_shape)
    
    def build(self, input_shape):
        self.model.build(input_shape=self.input_shape)
        self.model.compile(optimizer=self.optimizer, loss=self.loss, \
              metrics=[CategoricalAccuracy()])
        plot_model(self.model, to_file='cnn_model.png', show_shapes=True, show_layer_names=True)
        print("CNN model built!")
    
    def predict(self, input_image, verbose=1):
        out = self.model.predict(input_image, verbose=verbose)
        return self.one_hot(out)
    
    def one_hot(self, model_out):
        arg_max = np.argmax(model_out, axis=1).reshape(-1)
        out = np.eye(self.output_shape)[arg_max]
        return out # output = np.float32

    def fit(self, X, y, epochs=1, verbose=1):
        y_1h = np.eye(self.output_shape)[y.reshape(-1)]
        return self.model.fit(X, y_1h, epochs=epochs, verbose=verbose)
    
    def evaluate(self, X, y, verbose=1):
        y_1h = np.eye(self.output_shape)[y.reshape(-1)]
        return self.model.evaluate(X, y_1h, verbose=verbose)
````

In the cell below, we will also define a Keras Sequential network that will augment the data. The augmentations include randomly changing the brightness and contrast in the images. We will also define a helper function that will create a new dataset for training the model.

````python
def day_extraction_range(array, time_ids, times):
    if len(times) == 2:
        out0 = array[time_ids == times[0]]
        out1 = array[time_ids == times[1]]
        out = np.concatenate((out0, out1), axis=0)
        
    else:
        out0 = array[time_ids == times[0]]
        out1 = array[time_ids == times[1]]
        out2 = array[time_ids == times[2]]
        out = np.concatenate((out0, out1, out2), axis=0)
    return out

def make_aug_dataset(observations, actions, times):
    day_observations = day_extraction_range(observations, time_ids, times)
    day_actions = day_extraction_range(actions, time_ids, times)
    return day_observations, day_actions
````

Convolutional Neural Network ( CNN )
Dense Neural Network ( Dense NN )
